```python
import pandas as pd
df1 = pd.read_csv("hies_state.csv")
df2 = pd.read_csv("hh_lq_state.csv")
df = pd.merge(df1, df2, on='state')
df.head()
```

```python
df['exp_rate'] = df['expenditure_mean']/df['income_mean']*100
df['savings'] =(df['income_mean'] - df['expenditure_mean'])/df['income_mean']*100
df
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
# Set plot size
plt.figure(figsize=(10, 6))

# Create scatter plot
sns.scatterplot(x='income_mean', y='expenditure_mean', data=df)

# Set title and labels
plt.title('Scatter Plot of Income Mean vs. Expenditure Mean')
plt.xlabel('Income Mean')
plt.ylabel('Expenditure Mean')

# Show plot
plt.show()
```

```python
plt.figure(figsize=(8, 6))
plt.scatter(df['income_mean'], df['expenditure_mean'], s=df['households'] / 1000, alpha=0.5)
plt.title('Bubble Chart: Income vs Expenditure (Bubble size by population)')
plt.xlabel('Income Mean')
plt.ylabel('Expenditure Mean')
plt.show()
```

```python
# Set the plot size
plt.figure(figsize=(10, 6))

# Create the bar plot using seaborn
sns.barplot(x='state', y='households', data=df, palette='viridis')

# Set the title and labels
plt.title('Number of Households by State')
plt.xlabel('State')
plt.ylabel('Number of Households')

# Rotate x-axis labels for readability
plt.xticks(rotation=45)

# Display the plot
plt.show()
```
